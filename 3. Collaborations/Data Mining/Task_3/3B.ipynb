```python
X=boston_df.drop("Price",axis=1)
y=boston_df.Price
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
# train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=0)
```

```python
# The reason I plot it so that I can choose the suitbale range of alpha to plot median scores with respective alpha
#Associated with each alpha value is a vector of ridge regression coefficients, I store in a matrix coefs.
coefs_ridge = []
alphas_ridge = np.logspace(-2, 0, 20)
for i in alphas_ridge:
    ridge = Ridge(normalize=True).set_params(alpha = i).fit(X_train, y_train)
    coefs_ridge.append(ridge.coef_)
np.shape(coefs_ridge)
# 20 rows (one for each alphas) and 93 columns (one for each value of alpha)
# the coef estimates are expected smaller if L2 norm with large alpha , plot to see it 
ax = plt.gca()
ax.plot(alphas_ridge, coefs_ridge)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alphas_ridge')
plt.ylabel('coef_ridge')
```

```python
nb_folds = 10
cv = KFold(n_splits=nb_folds)
alphas = np.logspace(-1, 0.5, 20)
sr_mae_ridge ={}
sr_rmse_ridge ={}
# create loop alpha to run in Ridge model with required scoring and then compute average mean with alpha is index
for i in alphas:
    model = Ridge(alpha=i,fit_intercept=False,normalize=True).fit(X_train, y_train)
    results = cross_validate(model, X, y, cv=cv,
                                scoring=('neg_root_mean_squared_error','neg_median_absolute_error'),return_estimator=True)
    sr_mae_ridge[i]= pd.DataFrame(results).mean()['test_neg_median_absolute_error']
    sr_rmse_ridge[i]= pd.DataFrame(results).mean()['test_neg_root_mean_squared_error']

sr_mae_ridge
```

```python
sr_rmse_ridge
```

```python
fig = plt.figure(figsize=[9,5])
plt.plot(pd.Series(sr_mae_ridge), marker='o', label='MAE',color= 'red')
plt.plot(pd.Series(sr_rmse_ridge), marker='o', label='MSE',color='green')
plt.xlabel('alphas')
plt.ylabel('scores')
plt.legend()
plt.title('RIDGE - MAE vs MSE', fontsize=20)
```
